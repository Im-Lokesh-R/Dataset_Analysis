```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv('smart_watch_data.csv')
data
```

```python
"""
The tasks are:
1.	Check for null values and get information about the columns. Make data type conversions wherever required.
2.	Create a new column for “Total Minutes” to understand the overall daily activity levels of users.
3.	Find the relationship between calories burned and the total steps walked in a day.
4.	Distribution of steps using the histogram
5.	Bar plot of active vs. sedentary minutes
6.	Visualize the Average Activity Minutes Distribution with a Pie Chart.
"""
```

```python
data
```

```python
data.info()
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data.fillna(0,inplace=True)
```

```python
data.head()
```

```python
data['TotalMinutes']=data['VeryActiveMinutes']+data['FairlyActiveMinutes']+data['LightlyActiveMinutes']+data['SedentaryMinutes']
data.head()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='TotalSteps',y='Calories',data=data)
plt.title('relationship between totalsteps and calories burned')
plt.xlabel('total steps')
plt.ylabel('Calories burned')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data['TotalSteps'],bins=30,kde=True)
plt.title('Distibution of total steps')
plt.xlabel('total steps')
plt.ylabel('Frequency')
plt.show()
```

```python
data
```

```python
active_sedentary=data[['VeryActiveMinutes','FairlyActiveMinutes','LightlyActiveMinutes','SedentaryMinutes']].sum()

plt.figure(figsize=(10,6))
active_sedentary.plot(kind='bar')
plt.title('Active vs sedentaryminutes')
plt.xlabel('active minute')
plt.ylabel('sedentary minute')
plt.show()
```

```python
avg_minutes=data[['VeryActiveMinutes','FairlyActiveMinutes','LightlyActiveMinutes','SedentaryMinutes']].mean()

plt.figure(figsize=(8,8))
avg_minutes.plot(kind='pie',autopct='%1.1f%%',startangle=140)
plt.title('Avg active minutes distribution')
plt.ylabel('')
plt.show()
```
