```python
import pandas as pd
```

```python
"""The bowlers with the best bowling figures at the end of the matches"""
```

```python
t20_data=pd.read_csv('t20-world-cup-22.csv')
t20_data.head()
```

```python
t20_data['best bowler'].info()
```

```python
# here we have the best bowling figure 3-6 in this formate like wicktes and runs 

t20_data[['wickets','runs']]=t20_data['best bowling figure'].str.split('-',expand=True)
```

```python
t20_data['wickets']=pd.to_numeric(t20_data['wickets'])
t20_data['runs']=pd.to_numeric(t20_data['runs'])
t20_data.head()
```

```python
# lets find it 

t20_data=t20_data.sort_values(by=['wickets','runs'],ascending=[False,True])

best_bowling=t20_data.groupby('venue').first().reset_index()
best_bowling[['team1','team2','best bowler','wickets','runs','best bowling figure']]
```
